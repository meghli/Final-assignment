Getting data from "List of helicopter prison escapes" Wikipedia article

```python
url = 'https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes'
```

Printing first 3 rows of list

```python
data = data_from_url(url)
for row in data[:3]:
    print(row)
```

```python
index = 0
for row in data:
    data[index] = row[:-1]
    index +=1
print(data[:3])
```

```python
for row in data:
    date = fetch_year(row[0])
    row[0] = date
print(data[:3])
```

```python
min_year = min(data, key=lambda x: x[0])[0]
max_year = max(data, key=lambda x: x[0])[0]

years = []
for year in range(min_year, max_year + 1):
    years.append(year)
print(years)
```

```python
attempts_per_year = []
for year in years:
    attempts_per_year.append([year,0])
print(attempts_per_year)
```

```python
# Instruction 1 - for each row in data
for row in data:
    for year_attempt in attempts_per_year: 
        # Instruction 2 - nothing to do here
        year = year_attempt[0] # Instruction 3 - assign the year value in year_attempt to year
        if row[0] == year:
            year_attempt[1] += 1

# Instruction 4 - print the results
print(attempts_per_year)
```
